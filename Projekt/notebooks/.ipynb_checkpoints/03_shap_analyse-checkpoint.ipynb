```python
import shap
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import json
import warnings
```

```python
#Ziel: Modell erklärbar machen mit SHAP

# Modell und Daten laden
model = joblib.load('../ml/model/random_forest_diabetes.pkl')
df = pd.read_csv('../data/diabetes_binary_health_indicators_BRFSS2015.csv')

with open('../ml/model/selected_features.json', 'r') as f:
    selected_features = json.load(f)

X = df[selected_features]

print("Creating small subset for SHAP analysis (100 samples)...")
X_sample = X.sample(n=100, random_state=42)
print(f"SHAP Subset: {X_sample.shape}")

print("Calculating SHAP Explainer...")
start_time = time.time()

explainer = shap.TreeExplainer(model)
print(f"SHAP Explainer created in {time.time() - start_time:.1f} seconds")

# Test-SHAP für eine einzelne Vorhersage berechnen
print("\nTesting SHAP for single prediction...")
sample_idx = 0
sample_data = X_sample.iloc[[sample_idx]]

shap_values = explainer.shap_values(sample_data)
print(f"SHAP values type: {type(shap_values)}")
if isinstance(shap_values, list):
    print(f"SHAP list length: {len(shap_values)}")
    for i, val in enumerate(shap_values):
        print(f"  shap_values[{i}].shape: {np.array(val).shape}")
else:
    print(f"SHAP array shape: {np.array(shap_values).shape}")

# Einfache Feature Importance Tabelle erstellen
print("\n=== Feature Importance basierend auf SHAP ===")

# SHAP für alle Samples berechnen (kann dauern)
print("Calculating SHAP for all samples...")
shap_values_all = explainer.shap_values(X_sample)

# SHAP-Werte für Diabetes-Klasse extrahieren
if isinstance(shap_values_all, list) and len(shap_values_all) == 2:
    shap_diabetes = shap_values_all[1]  # Klasse 1: Diabetes
else:
    shap_diabetes = shap_values_all

# Falls 3D-Array: (n_samples, n_features, n_classes) → auf 2D reduzieren
if len(shap_diabetes.shape) == 3:
    shap_diabetes = shap_diabetes[:, :, 1] if shap_diabetes.shape[2] == 2 else shap_diabetes[:, :, 0]

# Absolute SHAP-Werte mitteln
mean_abs_shap = np.abs(shap_diabetes).mean(axis=0)

# Top 10 Features
top_indices = np.argsort(mean_abs_shap)[-10:][::-1]

print("\nTop 10 Features nach SHAP Importance:")
for i, idx in enumerate(top_indices, 1):
    print(f"{i:2}. {selected_features[idx]:20} {mean_abs_shap[idx]:.5f}")

# Einfacher Bar Plot für Top Features
plt.figure(figsize=(10, 6))
top_features = [selected_features[i] for i in top_indices]
top_values = [mean_abs_shap[i] for i in top_indices]

bars = plt.barh(range(len(top_features)), top_values, color='skyblue')
plt.yticks(range(len(top_features)), top_features)
plt.xlabel('Durchschnittlicher absoluter SHAP-Wert')
plt.title('Top 10 Features nach SHAP Importance')
plt.gca().invert_yaxis()

# Werte anzeigen
for bar, val in zip(bars, top_values):
    width = bar.get_width()
    plt.text(width * 1.01, bar.get_y() + bar.get_height()/2, 
             f'{val:.4f}', ha='left', va='center')

plt.tight_layout()
plt.savefig('../ml/results/shap_top_features.png', dpi=300, bbox_inches='tight')
plt.show()

# Explainer speichern
joblib.dump(explainer, '../ml/model/shap_explainer.pkl')

print("\nSHAP analysis completed!")
print("Plot saved: '../ml/results/shap_top_features.png'")
print("Explainer saved: '../ml/model/shap_explainer.pkl'")
```
