```python
import joblib
import pandas as pd
import numpy as np
import json
```

```python
# 4. Finale Pipeline Export für Backend

print("Loading model and data...")
model = joblib.load('../ml/model/random_forest_diabetes.pkl')
explainer = joblib.load('../ml/model/shap_explainer.pkl')

with open('../ml/model/selected_features.json', 'r') as f:
    selected_features = json.load(f)

df = pd.read_csv('../data/diabetes_binary_health_indicators_BRFSS2015.csv')

print(f"Model loaded")
print(f"Features: {len(selected_features)}")

def predict_diabetes_risk(input_features):
    """
    Gibt Diabetes-Risiko und SHAP-Erklärung zurück
    input_features: Dict mit Feature-Namen als Keys
    """
    try:
        # Features in richtiger Reihenfolge
        features_list = []
        for feature in selected_features:
            if feature in input_features:
                val = input_features[feature]
                # Zu Float konvertieren
                if isinstance(val, (list, np.ndarray)):
                    features_list.append(float(val[0]))
                else:
                    features_list.append(float(val))
            else:
                features_list.append(float(df[feature].median()))
        
        # Als DataFrame
        features_df = pd.DataFrame([features_list], columns=selected_features)
        
        # Vorhersage
        risk_probability = model.predict_proba(features_df)[0][1]
        
        # SHAP Werte für diese einzelne Vorhersage
        shap_values_single = explainer.shap_values(features_df)
        
        # Für binäre Klassifikation: shap_values_single ist eine Liste mit zwei Elementen (für jede Klasse)
        # Wir nehmen die Werte für die positive Klasse (Diabetes)
        if isinstance(shap_values_single, list) and len(shap_values_single) == 2:
            shap_vals = shap_values_single[1][0]  # Diabetes-Klasse, erste (einzige) Zeile
        else:
            shap_vals = shap_values_single[0]
        
        # Falls base_value enthalten ist, entfernen
        if len(shap_vals) == len(selected_features) + 1:
            shap_vals = shap_vals[:-1]
        
        # Feature Impact als Dictionary
        shap_dict = {}
        for i, feature in enumerate(selected_features):
            if i < len(shap_vals):
                shap_val = shap_vals[i]
                # Sicherstellen, dass es ein Skalar ist
                if isinstance(shap_val, (np.ndarray, list)):
                    shap_dict[feature] = float(shap_val[0])
                else:
                    shap_dict[feature] = float(shap_val)
            else:
                shap_dict[feature] = 0.0
        
        # Top 3 Features die Risiko erhöhen/verringern
        sorted_features = sorted(shap_dict.items(), key=lambda x: x[1], reverse=True)
        top_positive = sorted_features[:3]
        top_negative = sorted_features[-3:]
        
        return {
            'success': True,
            'risk_percent': round(float(risk_probability * 100), 1),
            'shap_values': shap_dict,
            'top_positive': [{'feature': f, 'impact': round(v, 4)} for f, v in top_positive],
            'top_negative': [{'feature': f, 'impact': round(v, 4)} for f, v in top_negative],
            'features_used': selected_features
        }
    
    except Exception as e:
        import traceback
        return {
            'success': False,
            'error': f"{str(e)}\n{traceback.format_exc()}"
        }


print("\nTesting prediction function...")

# Test 1: Medianwerte
print("\nTest 1: Mit Medianwerten")
test_input = {}
for feature in selected_features:
    test_input[feature] = float(df[feature].median())

result = predict_diabetes_risk(test_input)
if result['success']:
    print(f"Diabetes Risk: {result['risk_percent']}%")
    print("\nTop features INCREASING risk:")
    for item in result['top_positive']:
        print(f"  {item['feature']}: +{item['impact']:.4f}")
    print("\nTop features DECREASING risk:")
    for item in result['top_negative']:
        print(f"  {item['feature']}: {item['impact']:.4f}")
else:
    print(f"Error: {result['error']}")

# Test 2: Hohes Risiko
print("\nTest 2: Mit erhöhtem Risiko-Profil")
high_risk_input = {
    'HighBP': 1,
    'HighChol': 1,
    'BMI': 35,
    'PhysActivity': 0,
    'GenHlth': 4,
    'Age': 10  # Ältere Person
}

result2 = predict_diabetes_risk(high_risk_input)
if result2['success']:
    print(f"Diabetes Risk (hohes Risiko): {result2['risk_percent']}%")
else:
    print("Error")

final_pipeline = {
    'model': model,
    'explainer': explainer,
    'features': selected_features,
    'predict_function': predict_diabetes_risk,
    'feature_descriptions': {
        'HighBP': 'Bluthochdruck (0=Nein, 1=Ja)',
        'HighChol': 'Hoher Cholesterin (0=Nein, 1=Ja)',
        'BMI': 'Body Mass Index',
        'Smoker': 'Raucher (0=Nein, 1=Ja)',
        'Stroke': 'Schlaganfall in Vergangenheit (0=Nein, 1=Ja)',
        'HeartDiseaseorAttack': 'Herzerkrankung (0=Nein, 1=Ja)',
        'PhysActivity': 'Körperliche Aktivität (0=Nein, 1=Ja)',
        'Fruits': 'Obstkonsum (0=Nein, 1=Ja)',
        'Veggies': 'Gemüsekonsum (0=Nein, 1=Ja)',
        'HvyAlcoholConsump': 'Starker Alkoholkonsum (0=Nein, 1=Ja)',
        'GenHlth': 'Allgemeine Gesundheit (1=Exzellent, 5=Schlecht)',
        'MentHlth': 'Psychische Gesundheitstage (0-30 Tage)',
        'PhysHlth': 'Körperliche Gesundheitstage (0-30 Tage)',
        'Age': 'Altersgruppe (1=18-24, 13=80+)',
        'Education': 'Bildungsstand (1=Kein Abschluss, 6=College)',
        'Income': 'Einkommen (1=<10k$, 8=>75k$)'
    }
}

joblib.dump(final_pipeline, '../ml/model/final_pipeline.joblib')

# Frontend Config
frontend_config = {
    'features': selected_features,
    'default_values': {feature: float(df[feature].median()) for feature in selected_features},
    'feature_info': final_pipeline['feature_descriptions']
}

with open('../ml/model/frontend_config.json', 'w', encoding='utf-8') as f:
    json.dump(frontend_config, f, indent=2, ensure_ascii=False)

print("\nFinal pipeline exported successfully!")
```
